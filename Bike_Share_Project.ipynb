```python
#@title Data cleaning With SQL

## 1. Identify and Handle Missing Values:

SELECT
  COUNT(*) - COUNT(ride_id) AS missing_ride_id,
  COUNT(*) - COUNT(rideable_type) AS missing_ridable_type,
  COUNT(*) - COUNT(started_at) AS missing_started_at,
  COUNT(*) - COUNT(ended_at) AS missing_ended_at,
  COUNT(*) - COUNT(start_station_name) AS missing_start_station_name,
  COUNT(*) - COUNT(start_station_id) AS missing_start_station_id,
  COUNT(*) - COUNT(end_station_name) AS missing_end_station_name,
  COUNT(*) - COUNT(end_station_id) AS missing_end_station_id,
  COUNT(*) - COUNT(start_lat) AS missing_start_lat,
  COUNT(*) - COUNT(start_lng) AS missing_start_lng,
  COUNT(*) - COUNT(end_lat) AS missing_end_lat,
  COUNT(*) - COUNT(end_lng) AS missing_end_lng,
  COUNT(*) - COUNT(member_casual) AS missing_member_casual,
FROM
`fit-oasis-417114.Bike_share_case_study.tripdata_2024`;

## Handle missing values: There are only a small number of missing values therefor it can be filtered out.

CREATE OR REPLACE TABLE `fit-oasis-417114.Bike_share_case_study.tripdata_2024_cleaned` AS
SELECT * FROM `fit-oasis-417114.Bike_share_case_study.tripdata_2024`
WHERE ride_id IS NOT NULL
  AND started_at IS NOT NULL
  AND ended_at IS NOT NULL
  AND start_station_name IS NOT NULL
  AND start_station_id IS NOT NULL
  AND end_station_name IS NOT NULL
  AND end_station_id IS NOT NULL
  AND start_lat IS NOT NULL
  AND start_lng IS NOT NULL
  AND end_lat IS NOT NULL
  AND end_lng IS NOT NULL
  AND member_casual IS NOT NULL;

## This query creates a new table that excluded the all the Null values.

## 2. Address Data Type Issues:

## Verify data type:

SELECT * FROM `fit-oasis-417114.Bike_share_case_study.tripdata_2024` LIMIT 10;

## Correct Data Types: Use CAST() or PARSE_DATE(), PARSE_TIMESTAMP() to convert data types.

CREATE OR REPLACE TABLE `fit-oasis-417114.Bike_share_case_study.tripdata_2024_cleaned` AS
SELECT
  ride_id,
  rideable_type,
  PARSE_TIMESTAMP('%Y-%m-%d %H:%M:%E*S%Ez', CAST(started_at AS STRING)) AS started_at,
  PARSE_TIMESTAMP('%Y-%m-%d %H:%M:%E*S%Ez', CAST(ended_at AS STRING)) AS ended_at,
  start_station_name,
  start_station_id,
  end_station_name,
  end_station_id,
  start_lat,
  start_lng,
  end_lat,
  end_lng,
  member_casual
FROM
  `fit-oasis-417114.Bike_share_case_study.tripdata_2024`;

## 3. Remove any duplicate records:

SELECT ride_id, COUNT(*)
FROM `fit-oasis-417114.Bike_share_case_study.tripdata_2024`
GROUP BY ride_id
HAVING COUNT(*) > 1;

## After identifying the duplicte will now be removed.

CREATE OR REPLACE TABLE `fit-oasis-417114.Bike_share_case_study.tripdata_2024_cleaned` AS
SELECT * EXCEPT(row_num)
FROM(
  SELECT *, ROW_NUMBER() OVER (PARTITION BY ride_id ORDER BY started_at) AS row_num FROM `fit-oasis-417114.Bike_share_case_study.tripdata_2024`
)
WHERE row_num = 1;

## This query removed all the duplicate rows and kept only the first instances.

## 4. Handle outliers:

CREATE OR REPLACE TABLE `fit-oasis-417114.Bike_share_case_study.tripdata_2024_cleaned` AS
SELECT *
FROM `fit-oasis-417114.Bike_share_case_study.tripdata_2024`
WHERE TIMESTAMP_DIFF(ended_at, started_at, MINUTE) BETWEEN 1 AND 1440;

## This query removes rides that are less than a minute and more than a day.

## 5. Standardize text data:

CREATE OR REPLACE TABLE `fit-oasis-417114.Bike_share_case_study.tripdata_2024_cleaned` AS
SELECT
 ride_id,
 rideable_type,
 started_at,
 ended_at,
 TRIM(start_station_name) AS start_station_name,
 TRIM(end_station_name) AS end_station_name,
  start_station_id,
  end_station_id,
  start_lat,
  start_lng,
  end_lat,
  end_lng,
  member_casual
FROM
  `fit-oasis-417114.Bike_share_case_study.tripdata_2024`


## This query removes any whitespaces from string columns.
```

```python
#@title Tables for Tableau

## The following queries will be used to extract tables for use in Tableau for creating visualizations.

##1. Daily Ride Counts and Member/Casual Breakdown:

CREATE OR REPLACE TABLE `fit-oasis-417114.Bike_share_case_study.daily_rides_summary_cleaned` AS
SELECT
  EXTRACT(DATE FROM started_at) AS ride_date,
  COUNT(*) AS total_rides,
  COUNT(CASE WHEN member_casual = 'member' THEN 1 ELSE NULL END) AS member_rides,
  COUNT(CASE WHEN member_casual = 'casual' THEN 1 ELSE NULL END) AS casual_rides
FROM
  `fit-oasis-417114.Bike_share_case_study.tripdata_2024_cleaned`
GROUP BY
  ride_date
ORDER BY
  ride_date;

## 2. Hourly ride counts

  CREATE OR REPLACE TABLE `fit-oasis-417114.Bike_share_case_study.hourly_rides_summary_cleaned` AS
SELECT
  EXTRACT(HOUR FROM started_at) AS ride_hour,
  COUNT(*) AS total_rides
FROM
  `fit-oasis-417114.Bike_share_case_study.tripdata_2024_cleaned`
GROUP BY
  ride_hour
ORDER BY
  ride_hour;

## 3. Start station popularity

CREATE OR REPLACE TABLE `fit-oasis-417114.Bike_share_case_study.start_station_popularity_cleaned` AS
SELECT
  start_station_name,
  COUNT(*) AS start_count
FROM
  `fit-oasis-417114.Bike_share_case_study.tripdata_2024_cleaned`
WHERE start_station_name IS NOT NULL
GROUP BY
  start_station_name
ORDER BY
  start_count DESC;

## 4. End station popularity

CREATE OR REPLACE TABLE `fit-oasis-417114.Bike_share_case_study.end_station_popularity_cleaned` AS
SELECT
  end_station_name,
  COUNT(*) AS end_count
FROM
  `fit-oasis-417114.Bike_share_case_study.tripdata_2024_cleaned`
WHERE end_station_name IS NOT NULL
GROUP BY
  end_station_name
ORDER BY
  end_count DESC;

## 5. Rideable Type Usage

CREATE OR REPLACE TABLE `fit-oasis-417114.Bike_share_case_study.rideable_type_usage_cleaned` AS
SELECT
  rideable_type,
  COUNT(*) AS ride_count,
  COUNT(CASE WHEN member_casual = 'member' THEN 1 ELSE NULL END) AS member_count,
  COUNT(CASE WHEN member_casual = 'casual' THEN 1 ELSE NULL END) AS casual_count
FROM
  `fit-oasis-417114.Bike_share_case_study.tripdata_2024_cleaned`
GROUP BY
  rideable_type;

## 6. Average Ride Duration by User Type

CREATE OR REPLACE TABLE `fit-oasis-417114.Bike_share_case_study.avg_ride_duration_user_type_cleaned` AS
SELECT
  member_casual,
  AVG(TIMESTAMP_DIFF(ended_at, started_at, MINUTE)) AS avg_duration_minutes
FROM
  `fit-oasis-417114.Bike_share_case_study.tripdata_2024_cleaned`
GROUP BY
  member_casual;

## 7. Top start locations

CREATE OR REPLACE TABLE `fit-oasis-417114.Bike_share_case_study.top_10_start_locations` AS
SELECT
  start_station_name,
  start_lat,
  start_lng,
  COUNT(*) AS start_count
FROM
  `fit-oasis-417114.Bike_share_case_study.tripdata_2024_cleaned`
WHERE
  start_station_name IS NOT NULL AND start_lat IS NOT NULL AND start_lng IS NOT NULL
GROUP BY
  start_station_name, start_lat, start_lng
ORDER BY
  start_count DESC
LIMIT 10;

## 8. Top end locations

CREATE OR REPLACE TABLE `fit-oasis-417114.Bike_share_case_study.top_10_end_locations` AS
SELECT
  end_station_name,
  end_lat,
  end_lng,
  COUNT(*) AS end_count
FROM
  `fit-oasis-417114.Bike_share_case_study.tripdata_2024_cleaned`
WHERE
  end_station_name IS NOT NULL AND end_lat IS NOT NULL AND end_lng IS NOT NULL
GROUP BY
  end_station_name, end_lat, end_lng
ORDER BY
  end_count DESC
LIMIT 10;
```

Ridership peaks during the summer months (June-September), with September showing the highest usage. This peak is likely due to a combination of favorable weather, the start of the academic year, and local events.

**Rideable Type Preferences**:

Electric bikes are the most popular choice among both members and casual users, followed by classic bikes. Electric scooters show significantly lower usage.
